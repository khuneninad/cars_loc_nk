```python
def Location(query):
    query=query.title() #shows title of location or particular location
    df=pd.read_csv('/kaggle/input/cars-dataset/used_cars_data.csv')
    contain=df['Location'].str.contains(query) #Sorts list according to location query given
    return df[contain]
```

```python
Location('Mumbai')
```

```python
def fuel(query):
    query=query.title() #shows fuel type
    df=pd.read_csv('/kaggle/input/cars-dataset/used_cars_data.csv')
    contain=df['Fuel_Type'].str.contains(query) #shows list according to fuel type query given
    return df[contain]
```

```python
fuel('Petrol')
```

**MODE OF TRANSMISSION**

```python

def trans(query):
    query = query.title()
    df=pd.read_csv('/kaggle/input/cars-dataset/used_cars_data.csv')
    contain = df['Transmission'].str.contains(query)
    return df[contain]
```

```python
trans('Manual')
```

**COUNT NUMBER OF EACH MODE OF TRANSMISSION**

```python
count_trans = df['Transmission'].value_counts()
count_trans
```

**COUNT NUMBER OF EACH TYPES OF FUEL TYPE**

```python
count_fuel = df['Fuel_Type'].value_counts()
count_fuel
```

**SHOW BAR GRAPH FOR FUEL TYPE**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,3))
count_fuel.plot(kind = 'bar', color = 'orange')
plt.title('Distribution of Fuel Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```
